```python

def recursive_key_print(data, level=0):
    
    for k in data.keys():
        if type(data[k]) == h5py.Dataset:
                  
            print(level*" " +"* " + k)
        else:
            print(level*" " + "- " + k)
            recursive_key_print(data[k], level+1)
```

```python
with h5py.File(bb_output, "r") as bb_h5:    
    #recursive_key_print(bb_h5)
    bb_wave = bb_h5["lambda_wave"][...]
    bb_spectrum = bb_h5["alf_spectrum"][...]
    
with h5py.File(rs_output, "r") as rs_h5:    
    #recursive_key_print(rs_h5)
    rs_wave = rs_h5["lambda_wave"][...]
    rs_spectrum = rs_h5["alf_spectrum"][...]
    
```

```python
fig, ax = plt.subplots(1,1, figsize=(14,10))

ax.plot(bb_wave*1e6, bb_spectrum, label="BB")
ax.plot(rs_wave*1e6, rs_spectrum, label="Stellar")


ax.grid(True)

ax.set_xlim(0,4)
ax.legend()

ax.set_xlabel("wavelength [um]")
ax.set_ylabel("Flux [$J s^{-1} m^{-1}$]")

pass
```
